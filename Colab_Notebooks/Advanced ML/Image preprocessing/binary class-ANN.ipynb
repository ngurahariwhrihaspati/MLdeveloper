```python
import pandas as pd

df = pd.read_csv('citrus.csv')
df.head()
```

```python
df.info()
```

```python
df.name[df.name == 'orange'] = 0
df.name[df.name == 'grapefruit'] = 1
df.head()
```

```python
dataset = df.values #dataframe into array
dataset
```

```python
X = dataset[:,1:6] #kolom 1-5 attribute
y = dataset[:,0] #kolom 1 label
```

```python
from sklearn import preprocessing # normalization
min_max_scaler = preprocessing.MinMaxScaler()
X_scale = min_max_scaler.fit_transform(X) # transform only attribute, because label already binary class
X_scale 
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_scale, y, test_size=0.3)
```

```python
import numpy as np

y_train = y_train.astype(np.float32)
y_test = y_test.astype(np.float32)
```

```python
from keras.models import Sequential
from keras.layers import Dense

model = Sequential([
    Dense(32, activation='relu', input_shape=(5,)),
    Dense(32, activation='relu'),
    Dense(1, activation='sigmoid'),
])

model.compile(optimizer='sgd', #stochastic gradient descent --> good for small dataset
              loss='binary_crossentropy',
              metrics=['accuracy'])
```

```python
model.fit(X_train, y_train, epochs=100)
```

```python
model.evaluate(X_test, y_test) # 0 --> loss value, 1 --> accuracy value
```

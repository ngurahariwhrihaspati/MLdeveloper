```python
from sklearn import datasets
from sklearn.model_selection import train_test_split

iris = datasets.load_iris()
atribut = iris.data
label = iris.target

# seperate datasets into train & test dataset
X_train, X_test, y_train, y_test = train_test_split(
    atribut, label, test_size=0.2, random_state=1)
```

```python
from sklearn import tree

decision_tree = tree.DecisionTreeClassifier()
model_pertama = decision_tree.fit(X_train, y_train)
model_pertama.score(X_test, y_test)
```

```python
from sklearn.decomposition import PCA

#membuat objek PCA dengan 4 principal components
pca = PCA(n_components=4)

#apply pca into datasets
pca_attribute = pca.fit_transform(X_train)

#analyze attribute variance respectively
pca.explained_variance_ratio_
```

```python
# PCA dengan 2 principal component
pca = PCA(n_components = 2)
X_train_pca = pca.fit_transform(X_train)
X_test_pca = pca.fit_transform(X_test)
```

```python
#uji akurasi clasifier
model2 = decision_tree.fit(X_train_pca, y_train)
model2.score(X_test_pca, y_test)
```
